```python
import numpy as np
import matplotlib.pyplot as plt
```

### **Problema 1**
_Un estudio determinó que al seleccionar al azar adultos que poseen teléfonos inteligentes, el 54% de
ellos los usa estando en clase o en reuniones. Se quiere encontrar la probabilidad de que, seleccionando
al azar 8 personas con teléfono inteligente, exactamente 6 de ellas los utilicen en clase o reuniones._

```python

from scipy.stats import binom

# Graficando Binomial
N, p = 8, 0.54 # parametros de forma

binomial = binom(N, p) # Distribución
x = np.arange(binomial.ppf(0.01),
              binomial.ppf(0.99))
fmp = binomial.pmf(x) # Función de Masa de Probabilidad
pb = binom.pmf(6,8,0.54)

plt.plot(x, fmp, '--', color='k')
plt.vlines(x, 0, fmp, colors='blue', lw=5, alpha=0.5)
plt.hlines(pb, 0, 6, linestyles ="dashed", colors ="orange")
plt.title('Distribución Binomial')
plt.ylabel('probabilidad')
plt.xlabel('valores')
plt.show()
```

```python
print(f'''La probabilidad de que, seleccionando al azar 8 personas con teléfono inteligente y que exactamente 6 de ellas los utilicen en clase o reuniones
es de un {round(pb*100,3)}%''')
```

### **Problema 2**
_Durante un año reciente, una clínica registró 4221 nacimientos. Con este dato único, determinar la
probabilidad de que haya 15 nacimientos en 1 día. ¿Es poco frecuente este suceso?_

```python

from scipy.stats import poisson
# Graficando Poisson

mu =  (4221/365) # parametro de forma

poisson = poisson(mu) # Distribución
x = np.arange(poisson.ppf(0.01),
              poisson.ppf(0.99))
fmp = poisson.pmf(x) # Función de Masa de Probabilidad
pp = poisson.pmf(15)


plt.plot(x, fmp, '--', color='k')
plt.vlines(x, 0, fmp, colors='b', lw=5, alpha=0.5)
plt.hlines(pp, 0, 15, linestyles ="dashed", colors ="orange")
plt.title('Distribución Poisson')
plt.ylabel('probabilidad')
plt.xlabel('valores')
plt.show()
```

```python
print(f'''Existe un {round(pp*100,3)}% de probabilidades de que, en un día cualquiera, se produzcan exactamente 15 nacimientos.
El evento es poco probable, aunque en ningún caso imposible.''')
```
